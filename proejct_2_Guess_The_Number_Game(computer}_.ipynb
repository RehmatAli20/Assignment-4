```python
import random
EASY_LEVEL_TURNS=10
HARD_LEVEL_TURNS=5 # these are the constant value depends on the user input difficulty level of the game
def set_difficulty(level):
   if level.lower()=='easy': # here we are using .lower() if user put in caps then it will convert into lower then it will compare
    return EASY_LEVEL_TURNS
   elif level.lower()=='heard':
    return HARD_LEVEL_TURNS
   else: # why this ? becasue if user pass random value  then this code will not run , in order to not break the flow of code we are using this extra code
  # so as to user have the chance to again input valid type
    print("invalid input")
    level=input("choose level of diffuculty...Type 'easy or 'heard' ")
    set_difficulty(level) # here  again this function will run and in this time this function will take the above line code value as a parameter
def check_answer(guess, answer, attempts):
    if guess < answer:
        print("Your guess is too low")
        print("Guess a number:")
        return attempts - 1
    elif guess > answer:
        print("Your guess is too high")
        print("Guess a number:")
        return attempts - 1
    else:
        print(f"Your guess is right ... The answer was {answer}")
        return attempts  # Or you can return something else if needed

print("welcome to number guessing game")
def Game():
  print("let me think of a  number  between 1 to 20")

  answer=random.randint(1,20) # here random element will be generated by the computer and will be assign into the var answer
  print(answer)
  level=input("choose level of diffuculty...Type 'easy or 'heard' ") # here we are taking input from user which will pass as argument int function
  attemps=set_difficulty(level) # here we are calling the functions which is declare on the above line of code

  guess=0 # why guess=0 here ?  because when condition will check it will at first always true becasue  our random range is selected from 1-20  this will help to
  # run while loop at first
  while guess!=answer:
    print(f"you have {attemps} attemps to guess the number")

    guess=int(input("guess the number")) # taking input number as a guess from user
    if type(guess) !=int:
      print("please enter a valid....Type number [1,2,3,4,5....]")
      print("else you will lose your attemps")
      guess=int(input("guess the number"))
    else:
      attemps=check_answer(guess,answer,attemps) # calling the check number functions and passing the agrument
      if attemps==0: # will be check here is attemps are 0 or not if then this will return the fucntion
        print("you are out of guesses... you lose!")
        return
      elif guess!=answer:
        print("guess again")


Game();
```
